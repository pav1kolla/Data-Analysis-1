# C-4. What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this ?

```python
group=data.groupby(['brand','vehicleType'])['price'].agg('mean').unstack()
```

```python
group
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(group,annot=True,fmt='.2f',cbar=True)
plt.xlabel('brand')
plt.ylabel('vehicleType')
plt.title('avgs')
plt.show()
```
